```python
#Filling the injuries column using zero fill
df["Total.Fatal.Injuries"].fillna(0, inplace = True)
df["Total.Serious.Injuries"].fillna(0, inplace = True)
df["Total.Minor.Injuries"].fillna(0, inplace = True)
df["Total.Uninjured"].fillna(0, inplace = True)
```

```python
columns = [key for key, value in percentage_missing.items() if value<10]
df.dropna(subset = columns, inplace = True)
```

```python
#Fiilling the aircraft category column missing values with data from make and model to predict the category, if not unknown
def fill_nan_aircraft_category(df):
    df["Aircraft.Category"] = df.groupby("Make")["Aircraft.Category"].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))
    return df
fill_nan_aircraft_category(df)
```

THE NUMBER OF ACCIDENTS OCCURED THROUGH THE YEARS

```python
plt.figure(figsize=(12, 6))
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
accidents_per_year = df['Year'].value_counts().sort_index()
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

THE CATEGORY OF EACH  AIRCRAFT IN THE NUMBER OF ACCIDENTS THAT HAVE OCCURED

```python
	
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index)
plt.title('Distribution of Accidents by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()




```

THE MAKES WHICH CONTRIBUTE TO THE MOST ACCIDENTS


```python
plt.figure(figsize=(10, 6))
top_makes = df['Make'].value_counts().nlargest(10).index
sns.countplot(data=df[df['Make'].isin(top_makes)], x='Make', order=top_makes)
plt.title('Top 10 Makes with Most Accidents')
plt.xlabel('Make')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

THE NUMBER OF FATAL INJURIES TAKEN PLACE BY AIRCRAFTS

```python
	
plt.figure(figsize=(16, 6))
sns.boxplot(data=df, x='Aircraft.Category', y='Total.Fatal.Injuries')
plt.title('Box Plot of Fatal Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=45)
plt.show()
```
